### 4) Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pyodbc
```

```python
df = pd.read_excel(r"C:\Users\KOMAL\Downloads\amazon-sales-analytics.xlsx")
df.head(5)
```

### Data Cleaning

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["Order Date","EmailID","Product Name"], inplace=True)
```

```python
df.isnull().sum()
```

```python
df[['Country',"city","state"]] = df["Geography"].str.split(",",expand = True)
df.head(3)
```

### 1. What are the total Sales, Quantity, and Profit?

```python
Sales = df["Sales"].sum().round()
Qty = df["Quantity"].sum().round()
profit = df["Profit"].sum().round()
```

```python
print("The total sales is :",Sales)
print("The total quantity is :",Qty)
print("The total profit is :",profit)
```

### 2. Which are the Top 10 highest selling products?

```python
gr = df.groupby("Product Name").agg({"Quantity":"sum"}).round()
gr = gr.sort_values(by="Quantity",ascending=False).head(10).reset_index()
gr.head(10)
```

```python
plt.figure(figsize=(5,3))
ax = sns.barplot(data=gr, x="Product Name", y="Quantity",color="pink")
for i in ax.containers:
    ax.bar_label(i)
plt.xticks(rotation=90)
plt.title("Top 10 highest selling products")
plt.show()
```

### 3. Show top 5 categories which generates the highest Profit?

```python
grouped_data = df.groupby("Category").agg({"Profit": "sum"}).sort_values(by ="Profit",ascending=False).head(5).round().reset_index()
grouped_data
```

```python
Category = ["Copiers","Accessories","Binders","Paper","Phones"] 
Profit = [10000,20000,30000,40000,50000]
plt.pie(Profit, labels=Category, autopct='%1.1f%%', colors=['blue', "yellow",'red', 'green', 'orange'], 
wedgeprops={'edgecolor':'black', 'linewidth':0.4}) 
plt.title("CATEGORIES") 
plt.show()
```

### 4. State-wise Sales and Profit Analysis

```python
gr=df.groupby(["state"]).agg({"Sales":"sum","Profit":"sum"}).round(2).reset_index()
gr
ax= gr.set_index("state")[["Sales","Profit"]].plot(kind="bar",stacked=True)
```

### 5. Monthly Sales Trend

```python
df["Month"] = df["Order Date"].dt.month
```

```python
gr = df.groupby("Month").agg({"Sales":"sum"}).round(2).reset_index()
gr
```

```python
plt.figure(figsize=(5,3))
plt.plot(gr["Month"], gr["Sales"], color = "pink", marker = "o", linestyle ="--", markerfacecolor="red")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

```

### 6. Year-wise performance of Sales and Profit

```python
df["Year"] = df["Order Date"].dt.year
```

```python
gr = df.groupby("Year").agg({"Sales":"sum","Profit":"sum"}).round(2).reset_index()
gr
```

```python
ax= gr.set_index("Year")[["Sales","Profit"]].plot(kind="bar",stacked=False)
for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

### 7. Which customers placed the highest number of orders?

```python
gr = df.groupby("Order ID").agg({"EmailID":"count"}).round(2)
gr = gr.sort_values(by="EmailID",ascending=False).head(10).reset_index()
gr
```

```python
plt.bar(gr["Order ID"], gr["EmailID"], color="pink") 
plt.title("Highest number of orders")  
plt.xlabel("EmailID") 
plt.ylabel("Order ID")
plt.xticks(rotation=45)
plt.show()
```

### 8. Which products have negative profit (loss-making)?

```python
loss_products = df[df['Profit'] < 0].head(4)
```

```python
loss_products['Product Name'].unique()
```

### 9. What is the Average Order Value (AOV)?

```python
gr = df.groupby("Order ID").agg(Total_Sales=("Sales","sum")).round(2)
print("The Average Order value is:",gr["Total_Sales"].mean().round(1))
```

### 10. How many repeat customers do we have?

```python
gr= df.groupby("EmailID").agg(Total_Order=("EmailID","count")).sort_values(by = "Total_Order", ascending = False).reset_index()
gr
```

### 11. Which category sells the most quantity?

```python
gr = df.groupby('Category').agg({"Quantity":"count"}).round(2)
gr = gr.sort_values(by="Quantity",ascending=False).head(10).reset_index()
gr
```

### 12. Calculate Profit Margin % for each product

```python
gr = df.groupby("Product Name").agg({"Profit":"sum","Sales":"sum"})
gr
```

```python
df['Profit Margin %'] = (df['Profit'] / df['Sales']) * 100
df[['Product Name', 'Sales', 'Profit', 'Profit Margin %']].head()
```

### 13. Which is the most profitable month?


```python
df["Month"] = df["Order Date"].dt.month
```

```python
gr = df.groupby("Month").agg({"Profit":"sum"}).round(2)
gr = gr.sort_values(by="Profit",ascending=False).head(10).reset_index()
gr
```

### 14. What is the average shipping time?


```python
df.columns
```

```python
df["shipping time"]=(df['Ship Date']-df['Order Date']).dt.days
print("The average shipping time (in days) is",df["shipping time"].mean().round(2))
```

```python
sns.kdeplot(df["shipping time"])
plt.show()
```

### 15. Does Shipping Time impact Profit? (Correlation)

```python
df[["shipping time","Profit"]].corr()
```

#### Conclusion : The increase in shipping time is slightly affecting the profit i.e. profit is slightly reducing.
